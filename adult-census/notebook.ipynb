```python
import pandas as pd
data = pd.read_csv("adult.csv")
data = data.replace('?', 'Unknown')
data
```

```python
# explore dataset
data.education.value_counts()
```

```python
data.income.value_counts() # turn it into binary
```

```python
data = pd.concat([data.drop('workclass', axis=1), pd.get_dummies(data.workclass).add_prefix('workclass_').astype(int)], axis=1)
data = pd.concat([data.drop('occupation', axis=1), pd.get_dummies(data.occupation).add_prefix('occupation_').astype(int)], axis=1)
data = pd.concat([data.drop('native.country', axis=1), pd.get_dummies(data['native.country']).add_prefix('native.country_').astype(int)], axis=1)
data = data.drop('education', axis=1)
data = data.drop('fnlwgt', axis=1)
data = pd.concat([data.drop('relationship', axis=1), pd.get_dummies(data.relationship).add_prefix('relationship_').astype(int)], axis=1)
data = pd.concat([data.drop('race',axis=1),pd.get_dummies(data['race']).add_prefix('race_').astype(int)],axis=1)
data = pd.concat([data.drop('marital.status', axis=1), pd.get_dummies(data['marital.status']).add_prefix('marital.status_').astype(int)], axis=1)
```

```python
data
```

```python
# encode the remaining
data['sex'] = data['sex'].map({'Male': 1, 'Female': 0})
data['income'] = data['income'].map({'>50K': 1, '<=50K': 0})
```

```python
data
```

```python
# only numerical features now
# training 
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

train_data, test_data = train_test_split(data, test_size=0.3)
```

```python
train_data
```

```python
test_data
```

```python
# split into X and y
X_train = train_data.drop('income', axis=1)
y_train = train_data['income']
```

```python
X_test = test_data.drop('income', axis=1)
y_test = test_data['income']
```

```python
f = RandomForestClassifier()
f.fit(X_train, y_train)
```

```python
# evaluate on test data
f.score(X_test, y_test)
```

```python
# importances of features
imp = dict(zip(f.feature_names_in_, f.feature_importances_))
imp
```

```python
# hyperparameter tuning
from sklearn.model_selection import GridSearchCV

p_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [5, 10, 20],
    'min_samples_split': [2, 5],
    'max_features': ['sqrt', 'log2']
}

grid = GridSearchCV(estimator=RandomForestClassifier(), param_grid=p_grid, verbose=10)
grid.fit(X_train, y_train)
```

```python
grid.best_estimator_
```

```python
f = grid.best_estimator_
f.score(X_test, y_test)
```
